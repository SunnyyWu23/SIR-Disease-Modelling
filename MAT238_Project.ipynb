```python
import matplotlib.pyplot as plot # import the matplot library
from IPython.display import clear_output # import the animation library
import time, random
```

```python
# Parameters for the model
N = 100000 # Population size (CONSTANT)
v = 0.7 # rate of transition from Infected to Removed (CONSTANT)

B = 0.5
R = 0.5

# Set initial values, will be updated each time
S = 98500 #number of susceptible individuals at time t
I = 10 #number of infectious individuals at time t
R = 5 #number of removed individuals at time t
```

```python
# differential equations describing rate of change in the number of individuals in the three classes 

# instantiate the initial slopes

dSdt = (-B/N)*I*S # susceptible class
dIdt = (B/N)*I*S - v*I # infectious class
dRdt = v*I # removed class class

step_size = 0.1 # this determines the accuracy of plotted point
```

Use Euler's method for simulation


```python
plot.ion() # Start interactive mode

figure, axis = plot.subplots() # Extract the returned values

axis.set_xlim(0, 100)
axis.set_ylim(0, 100000)

t = 0 # time will be days elapsed 

while t < 100:

    axis.scatter(t,S, color='green')
    axis.scatter(t,I, color='red')
    axis.scatter(t,R, color='grey')

    # recompute slope for each loop
    dSdt = (-B/N)*I*S # susceptible class
    dIdt = (B/N)*I*S - v*I # infectious class
    dRdt = v*I # removed class class

    S = S + dSdt*step_size # Susceptible individuals model
    I = I + dIdt*step_size # Infectious individuals model
    R = R + dRdt*step_size # Removed individuals model 

    plot.pause(0.2)
    t+=1
```
